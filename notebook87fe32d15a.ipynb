# Complejidad de un algoritmo   

```python
import matplotlib.pyplot as plt 
import random
import time

def plot_times(times):
    fig, ax = plt.subplots(figsize=(12,8))

    ax.plot(times, linewidth = 2, color = 'r', linestyle = 'solid',
    marker = '>', markersize = 12, markerfacecolor = 'b',
    markeredgecolor = 'gray')

    ax.grid(True)
    plt.show

```

```python
plot_times([1, 3, 6, 7])
```

# Calcular maximo

Escribir una funcion `maximun()` que calcule el maximo de una lista de valores


```python
def maximun(values):
    answer = None
    for value in values:
        if answer == None or answer < value:
            answer = value
    return answer
```

```python
def random_inputs(length):
    return [random.randint(-1000, 10000) for _ in range(length)]
```

# Funcion min_max (values)

```python
def min_max (values):
    if not values:
        maximo = values[0]
        minimo = values[0]
    
    for value in values:
        if value > maximo:
            maximo = value
        if value < minimo:
            minimo = value
    
    return maximo, minimo
```
